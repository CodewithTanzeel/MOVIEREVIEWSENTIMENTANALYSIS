# **Essential Visualizations**

### Numerical Variables Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your data
df = pd.read_csv('IMDB_Processed.csv')

# Set style for better visualization
sns.set_style('whitegrid')

# Check for numerical columns
numerical_cols = df.select_dtypes(include=['number']).columns

if len(numerical_cols) > 0:
    # Histograms for all numerical columns
    print("Generating histograms for numerical columns...")
    df[numerical_cols].hist(figsize=(15, 10), bins=20)
    plt.suptitle('Histograms of Numerical Variables', y=1.02)
    plt.tight_layout()
    plt.show()

    # Boxplots for numerical columns
    print("\nGenerating boxplots for numerical columns...")
    plt.figure(figsize=(12, 6))
    sns.boxplot(data=df[numerical_cols].melt(), x='variable', y='value')
    plt.xticks(rotation=45)
    plt.title('Distribution of Numerical Variables')
    plt.tight_layout()
    plt.show()
else:
    print("No numerical columns found in the dataset. Cannot generate histograms or boxplots.")
```

### Categorical Variables Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.font_manager as fm

# Set up a font that supports more characters
plt.rcParams['font.family'] = 'DejaVu Sans'  # or another font that supports your characters


df = pd.read_csv('IMDB_Processed.csv')

# Clean text data by removing problematic characters
def clean_text(text):
    if pd.isna(text):
        return text
    return ''.join(char for char in str(text) if ord(char) < 128)

# Apply cleaning to all object columns
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].apply(clean_text)

# Plot categorical columns with error handling
cat_cols = df.select_dtypes(include=['object', 'category']).columns

for col in cat_cols:
    try:
        # Skip if column has too many unique values or is empty
        if df[col].nunique() > 50 or df[col].empty:
            print(f"Skipping {col} - too many unique values or empty")
            continue

        plt.figure(figsize=(10, 5))
        df[col].value_counts().plot(kind='bar')
        plt.title(f'Distribution of {col}')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
    except Exception as e:
        print(f"Error plotting {col}: {str(e)}")
```

### Text Length Analysis

```python
df = pd.read_csv('IMDB_Processed.csv')
# Add columns for text length metrics
df['review_length'] = df['review'].apply(len)
df['token_count'] = df['tokens'].apply(lambda x: len(x.split()) if isinstance(x, str) else 0)
df['lemma_count'] = df['lemmatized'].apply(lambda x: len(x.split()) if isinstance(x, str) else 0)

# Plot distributions
plt.figure(figsize=(15, 4))
plt.subplot(1, 3, 1)
sns.histplot(df['review_length'], bins=50)
plt.title('Review Length (Chars)')

plt.subplot(1, 3, 2)
sns.histplot(df['token_count'], bins=50)
plt.title('Token Count')

plt.subplot(1, 3, 3)
sns.histplot(df['lemma_count'], bins=50)
plt.title('Lemma Count')
plt.show()
```

### Word Frequency Analysis

```python
from wordcloud import WordCloud

# Raw reviews
wordcloud = WordCloud(width=800, height=400).generate(' '.join(df['review']))
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud: Raw Reviews')
plt.show()

# Lemmatized text
wordcloud = WordCloud(width=800, height=400).generate(' '.join(df['lemmatized']))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud: Lemmatized Text')
plt.show()
```

## N-gram Analysis

```python
from sklearn.feature_extraction.text import CountVectorizer

# Bigrams in lemmatized text
vectorizer = CountVectorizer(ngram_range=(2, 2))
bigrams = vectorizer.fit_transform(df['lemmatized'])
bigram_counts = pd.DataFrame(bigrams.sum(axis=0), columns=vectorizer.get_feature_names_out()).T.sort_values(0, ascending=False).head(10)

bigram_counts.plot(kind='barh', title='Top 10 Bigrams (Lemmatized)')
plt.show()
```
